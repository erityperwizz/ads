```python
import numpy as np
import pandas as pd
import io
import matplotlib.pyplot as plt

df = pd.read_csv('company_sales_data.csv')
df
```

```python
#line plot
df.plot(x="month_number",y="total_profit")
plt.title("Company profit per month")
```

```python
df.plot(x="month_number",y="total_profit",color="blue",linestyle="dotted",marker="o",linewidth=3, markersize=5)
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.legend(['Profit data of last year '], loc='lower right')
plt.title('Company Sales data of last year')
```

```python
# Extract data
month_number = df['month_number']
products = df.columns[1:-1]  # Exclude 'month_number' and 'total_profit'

# Plot each product separately
for product in products:
    plt.plot(month_number, df[product], marker='o', label=product)

# Plot settings
plt.title('Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales (in units)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
month_number = df['month_number']
toothpaste_sales = df['toothpaste']

# Plot
plt.scatter(month_number, toothpaste_sales, label='Toothpaste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend(loc="upper left")
plt.title("Toothpaste sales data")
plt.grid(True)
plt.show()
```

```python
facecream=df['facecream']

plt.bar([a-0.25 for a in month_number], facecream, label = 'total_profit', align='edge',color="pink")
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales Profit')
```

```python
faceWashSalesData=df['facewash']

plt.bar([a-0.25 for a in month_number], facecream, width= 0.25, label = 'Face Cream sales data', align='edge',color="Red")
plt.bar([a+0.25 for a in month_number], faceWashSalesData, width= -0.25, label = 'Face Wash sales data', align='edge',color="Blue")
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')
```

```python
profitList=df ['total_profit'].tolist()
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList, profit_range, label = 'Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
```

```python
bathingsoap=df['bathingsoap']

plt.bar(month_number, facecream, color='r')
plt.bar(month_number, toothpaste_sales, bottom=faceWashSalesData, color='y')
plt.bar(month_number, bathingsoap, bottom=toothpaste_sales, color='g')

plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(["Face Cream","Bathing Soap","Moisturizer"])
plt.title("Sales of each product")
plt.show()
```

```python
monthList  = df ['month_number'].tolist()

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(),
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```
